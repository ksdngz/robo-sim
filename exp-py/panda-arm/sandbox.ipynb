## 手先直線補間
軌跡の初期値$\bm{p}_0=[p_{x0}, p_{y0}, p_{z0}]^{T}$、終端値$\bm{p}_f=[p_{xf}, p_{yf}, p_{zf}]^{T}$とする直線上の目標値$\bm{p}_d$は以下となる。
$$
\bm{p}_d(s) = \bm{p}_0(1-s)+\bm{p}_fs \quad (s\in[0,1]) \tag{1}
$$
$s$を時間関数として定義し合成関数として表すと時間軌道が求まる。単純な例として、cos関数を使用して簡易的に曲線を用いる。
$$
s(t) = 1-\frac{1}{2}cos(\frac{\pi}{T}t) \quad (t\in[0,T])　\tag{2}
$$
代入すると、
$$
\bm{p}_d(t) = (\bm{p}_f - \bm{p}_0)(1-\frac{1}{2}cos(\frac{\pi}{T}t))+\bm{p}_0 \tag{3}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import planner.trajectoryProfiler as traj
from motioncon import motionControllerService as mcs

p0 = np.array([10,20,30])
pf = np.array([70,80,90])
T = 1000
traj: mcs.mr.mo.Trajectory = traj.MultiSignedProfiler.generateTraj(p0,pf,T)
plist : list[mcs.mr.mo.Point] = traj.pointList()
t = [p.time for p in traj.pointList()]
ps = [np.rad2deg(p.ps) for p in traj.pointList()]
#plt.plot(t, ps)

fig = plt.figure()

ax = fig.add_subplot(projection='3d')
x,y,z = [[p[i] for p in ps] for i in range(3)]
ax.scatter(x, y, z)

#ax2 = fig.add_subplot()
#ax2.plot(t, ps)

plt.show()
print('end')
```
